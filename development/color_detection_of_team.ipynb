```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
image_path = '../output_video/crroped_img.jpg'
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) 
```

```python
plt.imshow(image)
plt.show
```

```python
top_half_image = image[0:int(image.shape[0]/2),:]
plt.imshow(top_half_image)
plt.show()
```

```python
image_2d = top_half_image.reshape(-1,3)

kmeans = KMeans(n_clusters=2,random_state=0)
kmeans.fit(image_2d)

lables = kmeans.labels_

cluster_image = lables.reshape(top_half_image.shape[0], top_half_image.shape[1])

plt.imshow(cluster_image)
plt.show()
```

```python
corner_clusters = [cluster_image[0,0],cluster_image[0,-1],cluster_image[-1,0],cluster_image[-1,-1]]
non_player_cluster = max(set(corner_clusters), key=corner_clusters.count)
print(non_player_cluster)
```

```python
player_cluster = 1-non_player_cluster
print(player_cluster)
```

```python
kmeans.cluster_centers_[player_cluster]
```
